```python
store = data.groupby(['Store']).agg({'Fuel_Price':['mean','max','sum']})
store[:5]
```

```python
store2 = data.groupby(['Store']).agg({'Unemployment':['sum']})
```

```python
plt.figure(figsize = (15,8))
store[('Fuel_Price',  'sum')].plot()
store2[('Unemployment', 'sum')].plot()
plt.legend()
plt.xlabel('Store')
plt.ylabel('Factors')

plt.show()
```

Inference:

1.   The chart denotes the comparison of stores with various factors
2.   The factors taken into consideration are Fuel price and Unemployment



```python
avg_sales = data.groupby('Holiday_Flag')['Weekly_Sales'].mean().reset_index()
avg_sales.rename(columns={'Weekly_Sales': 'Average_Sales'}, inplace=True)
avg_sales['Holiday_Flag'] = avg_sales['Holiday_Flag'].replace({0: 'Non-Holiday Week', 1: 'Special Holiday Week'})

```

```python
plt.bar(x='Holiday_Flag', height='Average_Sales',data=avg_sales)
plt.xlabel('Holiday Week')
plt.ylabel('Ayg Sales')
```

Inference:


1.   The chart denotes the comparison of average sales of holiday week and non-holiday week
2.   It is observed that the average sales of holiday week is more compared to non-holiday week
